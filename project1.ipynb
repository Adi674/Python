```python
# Model Building Process
```

```python
X = df1.drop('Attrition',axis=1)
```

```python
y = df1['Attrition']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=2)
```

```python
print("Training data size",X_train.shape)
print("Testing data size",X_test.shape)
print("Training data answers size",y_train.shape)
print("Testing data answers size",y_test.shape)
```

```python
#Training the model using svm
from sklearn import svm
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train,y_train)
```

```python
#Model evaluation
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
y_pred_svm = classifier.predict(X_test)
```

```python
scores_accuracy=[]
```

```python
svmScore = accuracy_score(y_pred_svm,y_test)
```

```python
scores_accuracy.append(["SVM",svmScore])
```

```python
print("accuracy score of svm data:",svmScore*100)
```

```python
print(classification_report(y_pred_svm, y_test))
```

```python
confusion_matrix(y_test,y_pred_svm)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred_lr = lr.predict(X_test)
```

```python
lrscore = accuracy_score(y_pred_lr,y_test)
```

```python
scores_accuracy.append(["LR",lrscore])
```

```python
print("accuracy score of Logistic regression data:",lrscore*100)
```

```python
print(classification_report(y_pred_lr, y_test))
```

```python
confusion_matrix(y_test,y_pred_lr)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
best_accuracy=[]
for k in range(1,15):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    Y_pred=knn.predict(X_test)
    accuracy=accuracy_score(y_test,Y_pred)
    best_accuracy.append(accuracy)
    
plt.figure(figsize=(12,10))
plt.plot(range(1,15),best_accuracy,marker="o",markersize=10)
plt.ylabel("accuracy")
plt.xlabel("k value")
```

```python
#k value is greater in 14 
# and if we talk about consisitent value of k is 6 to 14 are cosnsistent
knn=KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred_knn=knn.predict(X_test)
```

```python
knnscore = accuracy_score(y_pred_knn,y_test)
```

```python
scores_accuracy.append(["KNN",knnscore])
```

```python
print("accuracy score of KNN data:",knnscore*100)
```

```python
print(classification_report(y_pred_knn, y_test))
```

```python
confusion_matrix(y_test,y_pred_knn)
```
